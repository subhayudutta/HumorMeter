```python
import pandas as pd
import numpy as np
```

**DATASET PREPROCESSING**

```python
df=pd.read_csv("/content/dataset.csv")
df.head()
```

```python
df['humor'] = df['humor'].map({True: 1, False: 0})
```

```python
df
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.dropna()
```

```python
X=df.drop('humor',axis=1)
y=df['humor']
```

```python
y.shape
```

**TEXT PREPROCESSING**

```python
import tensorflow as tf
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
```

```python
voc_size=5000
```

```python
import nltk
import re
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
corpus = []
for i in range(0, len(X)):
    review = re.sub('[^a-zA-Z]', ' ', X['text'][i])
    review = review.lower()
    review = review.split()

    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

**CONVERTING TEXT TO VECTOR**

```python
onehot_repr=[one_hot(words,voc_size)for words in corpus]
onehot_repr
```

```python
onehot_repr[1]
```

**EMBEDDING LAYER AND LSTM**

```python
sent_length=20
embedded_docs=pad_sequences(onehot_repr,padding='post',maxlen=sent_length)
print(embedded_docs)
```

```python
embedding_vector_features=40
model=Sequential()
model.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))
model.add(LSTM(100))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
print(model.summary())
```

```python
len(embedded_docs),y.shape
```

```python
X_final=np.array(embedded_docs)
y_final=np.array(y)
```

```python
X_final.shape,y_final.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_final, y_final, test_size=0.33, random_state=42)
```

**TRAINING THE MODEL**

```python
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=5,batch_size=64)
```

```python
y_pred=model.predict(X_test)
y_pred=np.where(y_pred > 0.6, 1,0)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

**SAVING THE MODEL AND MAKING PREDICTION**

```python
import pickle
pickle.dump(model,open('humorPred.pkl','wb'))
```

```python
pickled_model=pickle.load(open('humorPred.pkl','rb'))
```
